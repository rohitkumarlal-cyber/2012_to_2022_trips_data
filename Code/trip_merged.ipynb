```python
import pandas as pd
from pathlib import Path

master_trip = Path("C:/Users/axere/OneDrive/Desktop/CycleDA/cycle_data_13to22/Cycle_Data/Cleaned_data")
output_folder = Path("C:/Users/axere/OneDrive/Desktop/CycleDA/cycle_data_13to22/Cycle_Data/Merged_cleaned_data")

files = list(master_trip.glob("*.csv"))

dfs = [pd.read_csv(f) for f in files]

all_trips = pd.concat(dfs, ignore_index=True)

all_trips["usertype"] = all_trips["usertype"].replace({
    "Subscriber": "member",
    "Customer": "casual",
    "Dependent": "dependent"
})

all_trips = all_trips.drop_duplicates(subset=["unique_trip_id"])

output_file = output_folder / "all_Cleaned_trips.csv"
all_trips.to_csv(output_file, index =False )
```
